# print all printable characters included in ascii code and their hex values in ascending order.

```python
for i in range(0x21,0x7f):
  i_hex = hex(i)
  bytes_val = i.to_bytes(1, 'big')
  print(f'Hex: {i_hex} Char: {bytes_val.decode("ascii")}')
```

#  print all printable characters included in latin-1 code and their hex values in ascending order

```python
for i in range(0xA0,0x100):
  i_hex = hex(i)
  bytes_val = i.to_bytes(1, 'big')
  print(f'Hex: {i_hex} Char: {bytes_val.decode("latin-1")}')
```

# print all printable characters included in cp1252 code and their hex values in ascending order.

```python
for i in range(0x80, 0x100):  # Additional cp1252 characters
    i_hex = hex(i)
    bytes_val = i.to_bytes(1, 'big')
    try:
        print(f'Hex: {i_hex} Char: {bytes_val.decode("cp1252")}')
    except UnicodeDecodeError:
        print(f'Hex: {i_hex} Char: <Undefined in cp1252>')
```
