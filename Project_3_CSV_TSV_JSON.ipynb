```python
# pip is python install package commandline utility
# ! mark is need to run commandline tools
!pip install faker
import faker
import sys
import csv
```

```python
from faker import Faker

# select locale of fake data
fake = Faker('en_US')

# open a csv file to write
# utf-8 text encoding is selected to support emoji
with open('sample_data.csv', mode='w',encoding='utf-8') as sample_data:
  writer = csv.writer(sample_data)

  writer.writerow(['Name','State','Plate','SSN','Occupation','Emoji'])
  for _ in range(1000):
    writer.writerow([fake.name(), fake.state_abbr(), fake.license_plate(),fake.ssn(),fake.job(),fake.emoji()])
```

#CREATE A PANDAS DATA FRAME BY LOADING THE GENERATED FAKE TABLE

```python
import pandas as pd


df = pd.read_csv('/content/sample_data.csv')

df.head()
```

#DROP THE ROWS WHICH DOES NOT HAVE 'MO' IN STATE COLUMN

```python
df.drop(df[df['State'] != 'MO'].index, inplace = True)
```

```python
df.head()
```

#WRITE A FUNCTION WHICH TAKES IN A VALUE AND MODIFY IT TO THE WAY WE WANT

```python
def format_ssn(x):
    return ''.join(x.split('-'))
```

#PASS THE FUNCTION TO THE MAP FUNCTION TO AFFECT THE CHANGES

```python
df['SSN'] = df['SSN'].map(format_ssn)
df.head()
```

#JUST LIKE CSV, SAVE TO TSV BY GIVING TAB PLACEHOLDER AS A SEPARATOR

```python
df.to_csv('output_data.tsv', sep="\t")
```

#to_json IS USED TO SAVE IN JSON FORMAT

```python
df.to_json('output_data.json')
```
