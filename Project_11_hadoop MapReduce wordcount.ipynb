```python
%%writefile mapper.py
import sys
import io
import re
import nltk
nltk.download('stopwords',quiet=True)
from nltk.corpus import stopwords

punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
stop_words = set(stopwords.words('english'))

input_stream = io.TextIOWrapper(sys.stdin.buffer, encoding='latin1')

for line in input_stream:
    line = line.strip()
    line = re.sub(r'[^\w\s]', '', line)
    line = line.lower()

    for x in line:
        if x in punctuations:
            line = line.replace(x, " ")

    words = line.split()

    # Emit key-value pairs for each word with a count of 1
    for word in words:
        if word not in stop_words:
            print('%s\t%d' % (word, 1))

    # Additionally, emit a special key for the total count
    print('TotalWords\t%d' % len(words))
```

```python
%%writefile reducer.py

from operator import itemgetter
import sys

total_count = 0
current_word = None

# input comes from STDIN
for line in sys.stdin:
    # remove leading and trailing whitespace
    line = line.strip()
    line = line.lower()

    # parse the input we got from mapper.py
    word, count = line.split('\t', 1)
    try:
        count = int(count)
    except ValueError:
        # count was not a number, so silently
        # ignore/discard this line
        continue

    # accumulate the total count
    total_count += count

# output the total count to STDOUT
print('TotalWords\t%s' % total_count)
```
